```python
import selenium
import requests
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
import time
import pandas as pd
```

```python
driver = webdriver.Chrome('chromedriver',options=options)
```

```python
url = "https://shopee.co.id/samsung.official"
# url = "https://shopee.co.id/samsung.official?page=0&shopCollection=11354690"

driver.maximize_window()
driver.get(url)
driver.implicitly_wait(10)
```

```python
# MENGAMBIL SEMUA DATA YANG ADA DI URL
Nameclass = "shop-search-result-view__item"
content = driver.find_elements(By.CLASS_NAME,Nameclass)
len(content)
```

```python
# Get pagination number

total_page = driver.find_element(By.CLASS_NAME,"shopee-mini-page-controller__total").text
curent_page = driver.find_element(By.CLASS_NAME,"shopee-mini-page-controller__current").text
curent_page
```

```python
# driver.quit()
```

```python
# LOOPING DI SEMUA PAGE DAN MENGAMBIL DATA PRODUK



NAMA_PRODUK=[]
HARGA_PRODUK=[]
LINK_PRODUK=[]
PENJUALAN = []
for page in range(int(total_page)):
  
  if int(driver.find_element(By.CLASS_NAME,"shopee-mini-page-controller__current").text) > int(total_page):
    print("DONE")
    break
  else:
    # print(f"Halaman {page+1}")
    
    content = driver.find_elements(By.CLASS_NAME,Nameclass)
    LINK = [item.find_element(By.TAG_NAME,"a").get_attribute('href')  for item in content]
    NAMA= [item.find_element(By.CLASS_NAME,"_2mQnW2").text for item in content]
    HARGA = [item.find_element(By.CLASS_NAME,"_3TJGx5").text for item in content]
    TERJUAL = [item.find_element(By.CLASS_NAME,"_2Tc7Qg").text for item in content]
    LINK_PRODUK.extend(LINK)
    NAMA_PRODUK.extend(NAMA)
    HARGA_PRODUK.extend(HARGA)
    PENJUALAN.extend(TERJUAL)
    
    # print(content[0].find_element(By.CLASS_NAME, "_3Gla5X").text)
    

    next_btn = driver.find_element(By.CLASS_NAME,"shopee-mini-page-controller__next-btn").click()
```
